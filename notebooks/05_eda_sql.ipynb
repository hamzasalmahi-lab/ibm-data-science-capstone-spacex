# Notebook 5 — EDA with SQL

Ten SQL queries against the cleaned dataset loaded into SQLite. Documents site counts, payload aggregates and landing outcome rankings.

---


## Imports + load SQL magic


```python
%load_ext sql
import sqlite3, pandas as pd
con = sqlite3.connect('spacex.db')
%sql sqlite:///spacex.db
# Then ingest the cleaned CSV:
df = pd.read_csv('../data/dataset_part_2.csv')
df.to_sql('SPACEXTBL', con, if_exists='replace', index=False)
```

## Q1 — unique launch sites


```python
%sql SELECT DISTINCT Launch_Site FROM SPACEXTBL;
```

## Q2 — 5 records starting with 'CCA'


```python
%sql SELECT * FROM SPACEXTBL WHERE Launch_Site LIKE 'CCA%' LIMIT 5;
```

## Q3 — Total payload mass NASA (CRS)


```python
%sql SELECT SUM(PAYLOAD_MASS__KG_) AS total_payload_kg FROM SPACEXTBL WHERE Customer='NASA (CRS)';
```

## Q4 — Average F9 v1.1 payload


```python
%sql SELECT AVG(PAYLOAD_MASS__KG_) AS avg_payload_kg FROM SPACEXTBL WHERE Booster_Version='F9 v1.1';
```

## Q5 — First successful ground-pad landing


```python
%sql SELECT MIN(Date) AS first_success_date FROM SPACEXTBL WHERE Landing_Outcome='Success (ground pad)';
```

## Q6 — Drone-ship landings 4000–6000 kg


```python
%sql SELECT Booster_Version FROM SPACEXTBL WHERE Landing_Outcome='Success (drone ship)' AND PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000;
```

## Q7 — Mission outcome counts


```python
%sql SELECT Mission_Outcome, COUNT(*) AS n FROM SPACEXTBL GROUP BY Mission_Outcome;
```

## Q8 — Max-payload boosters


```python
%sql SELECT Booster_Version FROM SPACEXTBL WHERE PAYLOAD_MASS__KG_=(SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL);
```

## Q9 — 2015 failed drone-ship landings


```python
%sql SELECT substr(Date,6,2) AS month, Booster_Version, Launch_Site FROM SPACEXTBL WHERE substr(Date,1,4)='2015' AND Landing_Outcome='Failure (drone ship)';
```

## Q10 — Rank landing outcomes 2010-06-04 to 2017-03-20


```python
%sql SELECT Landing_Outcome, COUNT(*) AS n FROM SPACEXTBL WHERE Date BETWEEN '2010-06-04' AND '2017-03-20' GROUP BY Landing_Outcome ORDER BY n DESC;
```
