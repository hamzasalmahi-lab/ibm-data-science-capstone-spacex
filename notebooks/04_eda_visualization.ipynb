# Notebook 4 — EDA with Visualization

Pandas + seaborn exploration of the cleaned dataset across launch site, payload mass, orbit type, flight number and year.

---


## Imports


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load


```python
df = pd.read_csv('../data/dataset_part_2.csv')
```

## Flight Number vs. Launch Site


```python
sns.catplot(y='LaunchSite', x='FlightNumber', hue='Class',
            data=df, aspect=5)
```

## Payload Mass vs. Launch Site


```python
sns.catplot(y='LaunchSite', x='PayloadMass', hue='Class',
            data=df, aspect=5)
```

## Success rate per orbit (bar)


```python
orbit_success = df.groupby('Orbit')['Class'].mean().sort_values(ascending=False)
orbit_success.plot(kind='bar')
```

## Flight Number vs. Orbit Type


```python
sns.catplot(y='Orbit', x='FlightNumber', hue='Class', data=df, aspect=5)
```

## Payload Mass vs. Orbit Type


```python
sns.catplot(y='Orbit', x='PayloadMass', hue='Class', data=df, aspect=5)
```

## Yearly trend (line)


```python
df['Year'] = pd.to_datetime(df['Date']).dt.year
yearly = df.groupby('Year')['Class'].mean()
yearly.plot(marker='o')
```

## Build features for modelling


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit',
               'LaunchSite', 'Flights', 'GridFins', 'Reused',
               'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features_one_hot = pd.get_dummies(features,
                                  columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial'])
features_one_hot.to_csv('../data/dataset_part_3.csv', index=False)
```
