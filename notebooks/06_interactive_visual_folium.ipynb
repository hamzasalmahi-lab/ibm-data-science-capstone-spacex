# Notebook 6 — Interactive Visual Analytics with Folium

Build interactive maps of the four SpaceX launch sites with markers, outcome icons and proximity overlays.

---


## Imports


```python
import folium
import pandas as pd
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians
```

## Load + site centroids


```python
df = pd.read_csv('../data/spacex_launch_geo.csv')
launch_sites_df = df.groupby('Launch Site', as_index=False).first()[['Launch Site','Lat','Long']]
```

## Base map


```python
nasa_coord = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coord, zoom_start=5)
folium.Circle(nasa_coord, radius=1000, color='#000000', fill=True).add_to(site_map)
```

## Add site markers


```python
for _, row in launch_sites_df.iterrows():
    folium.Marker([row['Lat'], row['Long']],
                  icon=DivIcon(icon_size=(20,20), icon_anchor=(0,0),
                               html=f"<div style='font-size:12;color:#d35400'><b>{row['Launch Site']}</b></div>"
                              )).add_to(site_map)
```

## MarkerCluster of outcomes


```python
marker_cluster = MarkerCluster()
df['marker_color'] = df['class'].apply(lambda c: 'green' if c==1 else 'red')
for _, row in df.iterrows():
    folium.Marker([row['Lat'], row['Long']],
                  icon=folium.Icon(color='white', icon_color=row['marker_color'])
                 ).add_to(marker_cluster)
site_map.add_child(marker_cluster)
```

## Proximity analysis


```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6373.0
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon, dlat = lon2-lon1, lat2-lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    return R * 2 * atan2(sqrt(a), sqrt(1-a))

# Then drop folium.PolyLine + folium.Marker from each launch pad to
# coastline, railway, highway and the nearest city (e.g. Cocoa Beach).
```
