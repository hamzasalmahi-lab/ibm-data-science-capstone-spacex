# Notebook 3 — Data Wrangling

Engineer the binary landing-outcome label and produce the cleaned modelling dataset. Output: `data/dataset_part_2.csv`.

---


## Imports


```python
import pandas as pd
import numpy as np
```

## Load raw data


```python
df = pd.read_csv('../data/dataset_part_1.csv')
```

## Null-value inspection


```python
df.isnull().sum() / len(df) * 100
```

## Per-site and per-orbit counts


```python
df['LaunchSite'].value_counts()
df['Orbit'].value_counts()
```

## Outcome label engineering


```python
landing_outcomes = df['Outcome'].value_counts()
bad_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]])  # None None, False ASDS, etc.
landing_class = [0 if o in bad_outcomes else 1 for o in df['Outcome']]
df['Class'] = landing_class
```

## Headline statistic


```python
df['Class'].mean()  # ≈ 0.67 — about 67% of attempts succeed
```

## Export


```python
df.to_csv('../data/dataset_part_2.csv', index=False)
```
